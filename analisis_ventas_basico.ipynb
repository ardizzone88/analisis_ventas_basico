```python
import pandas as pd
import matplotlib.pyplot as plt

# Creamos un dataset simple
data = {
    'Producto': ['Laptop', 'Mouse', 'Monitor', 'Teclado', 'Audífonos'],
    'Ventas': [1500, 300, 800, 450, 600],
    'Cantidad': [5, 20, 10, 15, 12]
}

df = pd.DataFrame(data)
df.head()
```

```python
plt.bar(df['Producto'], df['Ventas'], color='skyblue')
plt.title('Ventas Totales por Producto')
plt.xlabel('Producto')
plt.ylabel('Ingresos (USD)')
plt.show()

#En este gráfico observamos que las Laptops representan la mayor fuente de ingresos, a pesar de tener un volumen de venta menor que los mouses.
```
